```python
data = data.fillna('')
```

```python
def clean_data(x):
        return str.lower(x.replace(" ", ""))
```

```python
new_features = ['title', 'director', 'cast', 'listed_in', 'description']
data = data[new_features]
```

```python
for new_features in new_features:
    data[new_features] = data[new_features].apply(clean_data)
```

```python
data.head()
```

```python
def create_cloumns(x):
    return x['title']+ ' ' + x['director'] + ' ' + x['cast'] + ' ' +x['listed_in']+' '+ x['description']
```

```python
data['new'] = data.apply(create_cloumns, axis=1)
```

```python
data.head()
```

```python
#Convert a collection of text documents to a matrix of token counts.
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity #Compute cosine similarity between samples in X and Y.
count = CountVectorizer(stop_words='english')#Terms that were ignored.
count_matrix = count.fit_transform(data['new'])
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
data=data.reset_index()
data
```

```python
indicate = pd.Series(data.index, index=data['title'])
```

```python
indicate.head()
```

```python
result = 0
def get_recommendations_new(title, cosine_sim):
    global result
    title=title.replace(' ','').lower()
    idx = indicate[title]

    # Get the pairwsie similarity scores 
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    indices = [i[0] for i in sim_scores]

    result =  data2['title'].iloc[indices]
    return result
```
